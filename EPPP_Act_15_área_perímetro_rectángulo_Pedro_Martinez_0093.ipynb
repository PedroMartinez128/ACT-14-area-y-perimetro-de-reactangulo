🔹 arearectangulo()
* Solicita la base y la altura al usuario.
* Aplica la fórmula base * altura.
* Imprime el resultado con formato .2f y la unidad unidades².

---


🔹 perimetrorectangulo()
* Pide nuevamente base y altura (cada función es independiente).
* Calcula 2 * (base + altura).
* Muestra el perímetro en unidades

---



🔹 menu()
* Utiliza while True: para mantener la ejecución activa hasta que el usuario decida salir.
* Presenta un encabezado visual con emojis y las 3 opciones disponibles.
* Evalúa la entrada con if / elif / else:
1 → Ejecuta arearectangulo()
2 → Ejecuta perimetrorectangulo()
3 → Muestra mensaje de despedida y rompe el bucle con break.
* Cualquier otro valor → Muestra advertencia y regresa al inicio del ciclo

---



```python
# 📐 Calculadora de Rectángulo - Nivel Principiante
# Autor: Pedro Martinez NL 34
# Formulas utilizadas :
# Área A = b × h
# Perímetro: P = 2 × (b + h)
# Fecha: 2026

def arearectangulo():
    """Calcula el área de un rectángulo"""
    base = float(input("📏 Ingresa la base del rectángulo: "))
    altura = float(input("📏 Ingresa la altura del rectángulo: "))
    area = base * altura
    print(f"🟩 El área del rectángulo es: {area:.2f} unidades²\n")

def perimetrorectangulo():
    """Calcula el perímetro de un rectángulo"""
    base = float(input("📏 Ingresa la base del rectángulo: "))
    altura = float(input("📏 Ingresa la altura del rectángulo: "))
    perimetro = 2 * (base + altura)
    print(f"🟦 El perímetro del rectángulo es: {perimetro:.2f} unidades\n")

def menu():
    """Muestra el menú principal y controla el flujo del programa"""
    while True:
        print("✨" * 25)
        print("      📐 MENÚ RECTÁNGULO 📐")
        print("✨" * 25)
        print("1️⃣  Calcular área del rectángulo")
        print("2️⃣  Calcular perímetro del rectángulo")
        print("3️⃣  Salir")

        opcion = input("👉 Selecciona una opción: ")

        if opcion == "1":
            arearectangulo()
        elif opcion == "2":
            perimetrorectangulo()
        elif opcion == "3":
            print("\n👋 ¡Hasta luego! Gracias por usar el programa. 🚀")
            print("Programa Realizado por Pedro Martinez NL = 34")
            break
        else:
            print("⚠️ Opción no válida. Por favor, elige un número del 1 al 3.\n")

# 🟢 Punto de entrada del programa
if __name__ == "__main__":
    menu()
```
